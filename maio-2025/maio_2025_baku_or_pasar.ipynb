````python
# Your work below
import pandas as pd
import re
````

````python
df = pd.DataFrame(train)
X = df[["text"]].copy()
y = df["class"].copy()
````

````python
y[y != 0] = 1
y.unique()
````

### Text Preprocessing

````python
!pip install malaya
````

````python
from malaya.tokenizer import Tokenizer
````

````python
!curl https://raw.githubusercontent.com/stopwords-iso/stopwords-ms/refs/heads/master/stopwords-ms.txt -O
````

````python
with open("stopwords-ms.txt", "r") as file:
    stopwords = [line.strip() for line in file]
````

````python
sanitizer = lambda x: list(filter(lambda w: w not in stopwords, tok.tokenize(x)))

tok = Tokenizer()
X1 = X.applymap(sanitizer)
X1
````

````python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(tokenizer=lambda x: x, token_pattern=None, lowercase=False)
res_tfidf = tfidf.fit_transform(X1.text)
res_tfidf
````

### Machine Learning

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import f1_score
````

````python
X_train, X_test, y_train, y_test = train_test_split(res_tfidf, y, random_state=42)
````

````python
sgd = SGDClassifier()
sgd.fit(X_train, y_train)
````

````python
y_pred = sgd.predict(X_test)
f1_score(y_test, y_pred)
````

### Make Predictions

````python
df1 = pd.DataFrame(test)
X_submit = df1[["text"]].copy()
X_submit
````

````python
X_submit = X_submit.applymap(sanitizer)
res_tfidf_submit = tfidf.transform(X_submit.text)
````

````python
y_pred_submit = sgd.predict(res_tfidf_submit)
y_pred_submit
````

````python
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', None)  # Auto-adjust width
pd.set_option('display.max_colwidth', None)
X_submit.assign(pred=y_pred_submit)
````

````python
answer = []
for i in range(len(y_pred_submit)):
    answer.append({"id": i, "class": int(y_pred_submit[i])})
answer
````

````python
post_answer({"solution": answer})
````

### Metrics

| Vectorizer | Model | Score |
| --- | --- | --- |
| `CountVectorizer` | `LogisticRegression` | 8.941176470588236 |
| `TfidfVectorizer` | `LogisticRegression` | 9.565217391304348 |

Key determining factors:
- Rare tokens are crucial; they appear more in *bahasa pasar*. TF-IDF prioritizes these tokens better than raw counts.
- Vectors are sparse. SVM handles this better as it focuses only on boundary cases (support vectors) and less sensitive to noise.
